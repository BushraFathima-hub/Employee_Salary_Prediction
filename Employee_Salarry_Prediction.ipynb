```python
X = data[['Years_at_Company','Satisfaction_Level','Average_Monthly_Hours']]
y = data["Salary"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
import joblib
joblib.dump(scaler,'scaler.pkl')
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_test
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error

def results(preddictions):
  print("Mean absolute error on model is {}",format(mean_absolute_error(y_test,preddictions)))
  print("Root mean squared error on model is {}",format(np.sqrt(mean_squared_error(y_test,preddictions))))
```

```python
from  sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
predictionlr = lr.predict(X_test)
```

```python
results(predictionlr)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
svrmodel = SVR()
```

```python
param_gridsvr={
    'C':[0.1,1,10],
    'epsilon':[0.1,0.2,0.5],
    'kernel':['linear','poly','rbf'],
    'degree':[2,3,4],
}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train, y_train.values.ravel())
```

```python
gridsvr.best_params_
```

```python
predictionssvr=gridsvr.predict(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
rfmodel=RandomForestRegressor()
```

```python
param_gridrfr={
    'n_estimators': [2,3,4,5,6],
    'max_depth':[5,10,15]
}
```

```python
gridrfr=GridSearchCV(estimator=rfmodel,param_grid=param_gridrfr,cv=5)
```

```python
gridrfr.fit(X_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
predictionsrfr=gridrfr.predict(X_test)
```

```python
results(predictionsrfr)
```
